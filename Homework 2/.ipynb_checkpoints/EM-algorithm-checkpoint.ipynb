# 2 Baseline: logistic regression

Каждого игрока логично закодировать one-hot представлением, а вес при этом слагаемом считать как skill игрока. Тогда вектор весов $\vec{w} = [powers] = [P_1, P_2, ... P_n, Q_1, Q_2, ... Q_m]$, где n - количество игроков, m - количество вопросов

```python
def apply_list_safe(id_, string):
    if string is None:
        return ['_'.join([str(id_), '0'])], [0]
    s_list = list(string)
    result = []
    for s in s_list:
        try:
            result.append(int(s))
        except:
            result.append(0)
    ids = ['_'.join([str(id_), str(i)]) for i in range(len(result))]
    return ids, result
```

```python
def prepare_data(data: pd.DataFrame):
    tournament = data.drop(columns=[NAME, DATE_START, DATE_END, TYPE, SEASON, 'orgcommittee', 'synchData', 'questionQty'])\
                        .explode([TEAMS, RESULTS, POSITION, TEAM_MEMBERS]).explode(TEAM_MEMBERS)
    tournament[[f'{ID}_q', RESULTS]] = tournament.apply(lambda row: apply_list_safe(row[ID], row[RESULTS]), axis=1,
                                                result_type='expand')
    tournament = tournament.explode([f'{ID}_q', RESULTS])
    tournament.dropna(inplace=True)
    return tournament
```

К тому же, мы будем иметь доступ к "весам" игроков и вопросов:

```python
class LogisticRegression(nn.Module):
    def __init__(self, n_feats):
        super().__init__()
        self.n_feats = n_feats
        
        self.linear = nn.Linear(n_feats, 1)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        return self.sigmoid(self.linear(x))
```

Теперь напишем реализацию обучения (m_step):

```python
def m_step(model, x, y, loss_function, opt, scheduler):
    opt.zero_grad()
    y_pred = model(x)
    loss = loss_function(y_pred, y)
    loss.backward()
    opt.step()
    scheduler.step()
    return loss.item()
```

```python
def train_lr(model, x, y, lr=1e-4, iterations=200):
    loss_function = nn.BCELoss()
    opt = torch.optim.AdamW(model.parameters(), lr=lr)
    scheduler = torch.optim.lr_scheduler.StepLR(opt, step_size=iterations // 10, gamma=0.5)
    losses = []
    
    fig, ax = plt.subplots()
    for i in tqdm(range(iterations)):
        loss_item = m_step(model, x, y, loss_function, opt, scheduler)
        losses.append(loss_item)
    
        #plot loss
        clear_output(wait=True)
        ax.plot(losses)
        ax.set_xlabel('step')
        ax.set_ylabel('loss')
        ax.set_title('Loss');
        display(fig)
    clear_output(wait=True)
```

Теперь найдем луших игроков:

```python
def top_players(model, ohe, n=30):
    player_skill = {}
    
    linear_weights = model.linear.weight.data[0].numpy()
    feature_names = ohe.get_feature_names()
    
    for idx, player_id in enumerate(feature_names):
        if player_id.startswith('x0_'):
            player_skill[int(player_id[3:])] = linear_weights[idx]
            
    top = sorted(
        [(player_id, power) for player_id, power in player_skill.items()],
        reverse=True, key=lambda x: x[1]
                )[:n]
    
    dict_top = defaultdict(list)
    
    for place, (player_id, power) in enumerate(top, 1):
        dict_top['id'].append(player_id)
        dict_top['skill'].append(power)
        dict_top['place'].append(place)
        dict_top[NAME].append(players_pd.loc[player_id][NAME])
        dict_top[SURNAME].append(players_pd.loc[player_id][SURNAME])
    dict_top = pd.DataFrame(dict_top)
    return dict_top
```

# 3. Рейтинг-система

Задание:

Качество рейтинг-системы оценивается качеством предсказаний результатов турниров. Но сами повопросные результаты наши модели предсказывать вряд ли смогут, ведь неизвестно, насколько сложными окажутся вопросы в будущих турнирах; да и не нужны эти предсказания сами по себе

Критерии:
- предложите способ предсказать результаты нового турнира с известными составами, но неизвестными вопросами, в виде ранжирования команд;
- в качестве метрики качества на тестовом наборе давайте считать ранговые корреляции Спирмена и Кендалла (их можно взять в пакете scipy) между реальным ранжированием в результатах турнира и предсказанным моделью, усреднённые по тестовому множеству турниров.

Решение:

Будем считать, что сила команды - это вероятность ответить на средний вопрос по турниру хотя бы 1 ее члена:

<center>$team skill = 1 - \prod_{i=1}^N (1 - \sigma(P_i + Q_{tournier mean} + bias)$</center>, где $P_i$ - skill игрока i, N - количество игроков в команде, $Q_{tourniermean}$ - средняя сложность вопроса в турнире

При этом будем считать, что $P_{unknown} = \frac{1}{n}\sum_{i}^{n} P_i$, где n - общее количество уже известных игроков

```python
class RankingEstimator(BaseEstimator):
    def __init__(self, model, ohe):
        self.PLAYER_SKILL = "player_skill"
        self.TOURNIER_MEAN_Q = 'tournier_mean_q'
        self.PLAYERS_PROB = 'players_prob'
        
        self.player_skill = {}
        self.tournier_mean_q = defaultdict(list)
        self.linear_weights = model.linear.weight.data[0].numpy()
        self.feature_names = ohe.get_feature_names()
        
        for idx, player_id in enumerate(self.feature_names):
            if player_id.startswith('x0_'):
                self.player_skill[int(player_id[3:])] = self.linear_weights[idx]
            else:
                self.tournier_mean_q[int(player_id.split('_')[1])].append(self.linear_weights[idx])
        self.tournier_mean_q = {t_id: np.mean(q) for t_id, q in self.tournier_mean_q.items()}
         
        self.mean_player_skill = sum(self.player_skill.values()) / len(self.player_skill)
        self.mean_tournier_q = sum(self.tournier_mean_q.values()) / len(self.tournier_mean_q)
        self.bias = model.linear.bias.data[0].numpy()
        
    def fit(self, data: pd.DataFrame):
        data = data.copy()
        data[self.PLAYER_SKILL] = data[TEAM_MEMBERS].map(self.player_skill)
        data[self.PLAYER_SKILL].fillna(self.mean_player_skill, inplace=True)
        
        data[self.TOURNIER_MEAN_Q] = data[ID].map(self.tournier_mean_q)
        data[self.TOURNIER_MEAN_Q].fillna(self.mean_tournier_q, inplace=True)
        
        data[self.PLAYERS_PROB] = data\
                .apply(lambda row: self.sigmoid(row[self.PLAYER_SKILL] + row[self.TOURNIER_MEAN_Q] + self.bias),
                       axis=1)
        
        ratings = data.groupby([ID, TEAMS])[self.PLAYERS_PROB].apply(lambda x: np.prod(1 - x))
        real_position = data.groupby([ID, TEAMS])[POSITION].first()
        
        result = pd.concat([ratings, real_position], axis=1)
        result.sort_values([ID, self.PLAYERS_PROB], ascending=[True, True], inplace=True)
        
        return result
    
    def predict(self, result: pd.DataFrame):
        spearman_rho = result.groupby(ID).apply(lambda x: self.easy_spearman(x[POSITION], x[self.PLAYERS_PROB])).mean()
        kendall_rho = result.groupby(ID).apply(lambda x: kendalltau(x[POSITION], x[self.PLAYERS_PROB]).correlation
        ).mean()
        
        return spearman_rho, kendall_rho
    
    def fit_predict(self, data: pd.DataFrame, return_data=False):
        result = self.fit(data)
        spearman_rho, kendall_rho = self.predict(result)
        if return_data:
            return result, spearman_rho, kendall_rho
        else:
            return spearman_rho, kendall_rho
        
    @staticmethod
    def sigmoid(x):
        return 1 / (1 + np.exp(-x))
    
    @staticmethod
    def easy_spearman(x, y):
        try:
            return spearmanr(x, y).correlation
        except Exception:
            return 1
```

```python
ranking_est = RankingEstimator(model, ohe)
```

```python
result, spearman_rho, kendall_rho = ranking_est.fit_predict(data_test, True)
```

- разработайте EM-схему для обучения этой модели, реализуйте её в коде;
- обучите несколько итераций, убедитесь, что целевые метрики со временем растут (скорее всего, ненамного, но расти должны), выберите лучшую модель, используя целевые метрики.


**Important**: Схема ниже придумана не мной, но подсказана Александром Тихомировым из набора MADE 20-21, за что ему спасибо, я очень долго не мог разобраться

Предположения:

1) При ответе 1 игрока команды на вопрос на вопрос автоматически отвечают все игроки команды: $p(team_i = 1|player_j = 1) = 1$ </br>
2) При условии неправильного ответа команды на вопрос ни один игрок этой команды не ответил на этот вопрос: $p(player_j = 1 | 
team_i = 0) = 0$</br>
3) Выполняется условие независимости ответов игроков на вопросы

Теперь мы будем предсказывать не такую вероятность:

<center> $p(player_j = 1) = 1$ </center>

А такую:

<center>$p(player_{j} = 1 | team_i = 1)$, где $j \in team_i$ </center>

По теореме Байеса:

<center>$p(player_{j} = 1 | team_i = 1) = \frac{p(team_{i} = 1 | player_j = 1)p(player_j = 1)}{p(team_i = 1)}$</center>

Это обновление вероятностей, которое представляет как раз Expectation-шаг

- $p(team_i = 1|player_j = 1) = 1$ по 1 предположению
- $p(player_j = 1) = \sigma(P_i + Q_{tournier mean} + bias)$
- $p(team_i = 1) = team skill = 1 - \prod_{i=1}^N (1 - \sigma(P_i + Q_{tournier mean} + bias)$

Тогда:

<center>$p(player_{j} = 1 | team_i = 1) = \frac{\sigma(P_i + Q_{tournier mean} + bias)}{1 - \prod_{i=1}^N (1 - \sigma(P_i + Q_{tournier mean} + bias)}$</center>

```python
def e_step(data: pd.DataFrame,
           new_predictions):
    print('1')
    data['y_pred'] = new_predictions
    # по второму предположению
    print('2')
    data.loc[data[RESULTS] == 0, 'y_pred'] = 0
    #знаменатель дроби
    index_where_1 = data[RESULTS] == 1
    print('3')
    denom = data.loc[index_where_1].groupby([TEAMS, 'id_q'])["y_pred"].transform(lambda x: 1 - np.prod(1 - x.values))
    print('4')
    data.loc[index_where_1, 'y_pred'] = data.loc[index_where_1, 'y_pred'] / denom
    print('5')
    return data
```

```python
def train_EM(model, ohe, x, data, iterations=50):
    data = data.copy()
    spearman = []
    kendall = []
    
    model.eval()
    predictions = model(x).detach().numpy().ravel()
    
    fig, ax = plt.subplots(1, 2, figsize=(15, 8))
    for i in tqdm(range(iterations)):
        model.train()
        data = e_step(data, predictions)
        y = torch.FloatTensor(data['y_pred'].values).view(-1, 1)
        
        print('resetting parameters')
        n_iter = 100
        #model.linear.reset_parameters()
        loss_function = nn.BCELoss()
        opt = torch.optim.AdamW(model.parameters(), lr=0.5)
        scheduler = torch.optim.lr_scheduler.StepLR(opt, step_size=n_iter // 10, gamma=0.5)
        
        for j in tqdm(range(n_iter)):
            m_step(model, x, y, loss_function, opt, scheduler)
        model.eval()
        predictions = model(x).detach().numpy().ravel()
        
        print('Ranking est...')
        ranking_est = RankingEstimator(model, ohe)
        spearman_rho, kendall_rho = ranking_est.fit_predict(data_test, False)
        
        spearman.append(spearman_rho)
        kendall.append(kendall_rho)
        
        clear_output(wait=True)
        print(f"Epoch: {i + 1}")
        print(f"Spearman: {spearman_rho}")
        print(f"Kendall: {kendall_rho}")
        
        ax[0].plot(spearman)
        ax[0].set_xlabel("Spearman rho:")
        ax[1].plot(kendall)
        ax[1].set_xlabel("Kendall rho: ")
        display(fig)
```
