```python
sigma_noise = np.std(y_train - model.predict(X_train))
```

```python
def bayesian_update(mu, sigma, x_matrix: np.ndarray, y: np.ndarray, sigma_noise=.25):
    sigma_n = np.linalg.inv(np.linalg.inv(sigma)+ (1 / (sigma_noise ** 2)) * np.matmul(np.transpose(x_matrix), x_matrix) )
    mu_n = np.matmul(sigma_n, np.matmul(np.linalg.inv(sigma), np.transpose(mu)) + (1 / (sigma_noise ** 2)) * np.matmul(np.transpose(x_matrix), y ) )
    return mu_n, sigma_n
```

```python
cur_mu, cur_sigma = bayesian_update(cur_mu, cur_sigma, X_train, y_train, sigma_noise=sigma_noise)
Z = multivariate_normal.pdf(pos, mean=cur_mu, cov=cur_sigma)
```

# 3.а Сигмоидальный вид функции числа заболевших

Выпишем сначала cdf для нормального распределения и рассмотрим его:

$Ф(x) = const\int_{-\inf}^{x} e^{-\frac{t^2}{2}}dt$

Здесь $t = w_2z^2 + w_1z + w_0$

$y = \int_{-\inf}^{x} e^{-(w_2z^2 + w_1z + w_0)}dz$

Понятно, что аналитически это нельзя посчитать. Но можно посчитать производную по y, это будет просто pdf:

$y' = e^{-w_2x^2 + w_1x + w_0}$

$ln(y') = w_2x^2 + w_1x + w_0$ - уже линейная фунция

Здесь можно сказать, что y' это просто новые случаи, то есть аналог производной

Построим pdf стандартного нормального распределения (пока одномерного) и посмотрим на него:

```python
plt.plot(stats.norm.pdf(np.linspace(-8, 8, 200), loc=0, scale=1));
```

Хорошо, как же теперь получить обратно y? Просто численно посчитать интеграл (кумулятивно посчитать значения функции)

$y_i = cumsum_0^i (w_2x_i^2 + w_1x_i + w_0)$

Мы получили план для реализации такой регрессии, осталось его воплотить:

```python
# Посэмплируем прямые и выдадим статистики по предсказаниям
def sample_statistics(mu, sigma, X, n=20):
    my_w = np.random.multivariate_normal(mu, sigma, n)
    res = np.zeros((n, X.shape[0]))
    for i,w in enumerate(my_w):
        res[i,:] = np.cumsum(np.exp(X @ w))
    return res
```

```python
# Нарисуем результат
def plot_predictions(xs, X, y_true, mu, x, points):
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111)
    ax.plot(xs, y_true, label="Правильный ответ")
    res = np.cumsum(np.exp(X @ cur_mu))
    ax.plot(xs, res, color="red", label="MAP гипотеза")
    ax.fill_between(xs, res - sigma_noise, res + sigma_noise, color=palette[1], alpha=.3, label="+- дисперсия шума")
    ax.fill_between(xs, np.mean(x, axis=0) - np.std(x, axis=0), np.mean(x, axis=0) + np.std(x, axis=0), color=palette[5], alpha=.2, label="+- дисперсия предсказаний")
    if points:
        ax.scatter(points[0], points[1], marker='*', s=200)
    ax.legend(fontsize=legend_fontsize)
```

```python
all_days = np.append(train.index.values, test.index.values) + 1
x = sample_statistics(cur_mu, cur_sigma, X_train, n=1000)
plot_predictions(train.index.values + 1, \
                 X_train,
                 train[TOTAL_CASES].values,
                 cur_mu, x, None)
```

Понятно, опять же, что, чем дальше, тем хуже будут предсказания

```python
x = sample_statistics(cur_mu, cur_sigma, X, n=1000)
plot_predictions(all_days, \
                 X,
                 np.append(train[TOTAL_CASES].values, test[TOTAL_CASES].values) + 1,
                 cur_mu, x, None)
```
